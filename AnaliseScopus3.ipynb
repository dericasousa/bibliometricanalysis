- Divisão de títulos e resumos em palavras ou frases (tokens);

#### Função para pré processamento de textos:

```python
def preprocess_text(text):
    # Verifica se o texto é NaN e trata-o adequadamente
    if pd.isna(text):
        return ''
    
    # Remove caracteres especiais, pontuações, e normaliza o texto
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)  # Remove caracteres especiais
    text = unicodedata.normalize('NFKD', text).encode('ASCII', 'ignore').decode('utf-8')  # Remove acentos
    
    # Detecta o idioma do texto
    lang = detect(text)
    
    # Tokenização: Divide o texto em palavras ou frases (tokens)
    tokens = word_tokenize(text.lower())  # Converte o texto em minúsculas durante a tokenização
    
    # Remoção de Stopwords: Elimina palavras comuns que não agregam muito significado
    if lang == 'en':
        stop_words = set(stopwords.words('english'))  # Lista de stopwords em inglês
    elif lang == 'pt':
        stop_words = set(stopwords.words('portuguese'))  # Lista de stopwords em português
    else:
        stop_words = set()  # Se o idioma não for reconhecido, não remove stopwords
    
    filtered_tokens = [word for word in tokens if word not in stop_words]
    
    # Stemming ou Lemmatization: Reduzir palavras às suas formas base
    if lang == 'en':
        stemmer = PorterStemmer()  # Stemmer (uso do stemming)
        lemmatizer = WordNetLemmatizer()  # Lemmatizer (uso da lematização)
    elif lang == 'pt':
        stemmer = RSLPStemmer()  # Stemmer para o português
        lemmatizer = WordNetLemmatizer()  # Lemmatizer (uso da lematização)
    else:
        stemmer = None
        lemmatizer = None
    
    if stemmer:
        stemmed_tokens = [stemmer.stem(word) for word in filtered_tokens]
    else:
        stemmed_tokens = filtered_tokens
    
    if lemmatizer:
        lemmatized_tokens = [lemmatizer.lemmatize(word) for word in filtered_tokens]
    else:
        lemmatized_tokens = filtered_tokens
    
    return ' '.join(lemmatized_tokens)  # Retorna o texto pré-processado como uma única string
```

```python
def extrair_universidades(dataFrame):
    institution_type = ['University', 'Institute']
    retorno = []
    for x in dataFrame[0]:
        for tipo in institution_type:
            if tipo in x:
                retorno.append(x)
    return retorno          
        
```

```python
def extrai_nomes(linha):
        ret = []
        nomes = [nome.strip() for nome in linha.split(',')]
        for i in range(len(nomes)):
            #print(nomes[i])
            if 'University' in nomes[i]:
                if nomes[i] not in ret:
                    ret.append(nomes[i])
            if 'Institute' in nomes[i]:
                if nomes[i] not in ret:
                    ret.append(nomes[i])        
        return ret
```

```python
def extrair_universidades_works(dataFrame):
    retorno = []
    for x in dataFrame['Affiliations']:
        unis = extrai_nomes(x)
        retorno.append(unis)
    return retorno
```

## Tipos de Dados de Sensores:

```python
from collections import Counter
from nltk.tokenize import word_tokenize
```

```python
# Função para contar a frequência de termos em um texto
def count_terms(text, terms):
    tokens = word_tokenize(text.lower())
    term_count = Counter(tokens)
    term_count = {term: term_count[term] for term in terms if term in term_count}
    return term_count
```

```python
# Função para criar um gráfico de barras com base em uma lista de termos
def generate_bar_chart(df, column, terms, title):
    df['Term_Count'] = df[column].apply(lambda x: count_terms(x, terms))
    total_term_counts = Counter()
    
    for term_count in df['Term_Count']:
        total_term_counts.update(term_count)
    
    if not total_term_counts:
        print("Nenhum dos termos está presente nos textos.")
        return
    
    term_freq = total_term_counts.items()
    term_freq = sorted(term_freq, key=lambda x: x[1], reverse=True)
    
    terms, frequencies = zip(*term_freq)
    
    plt.figure(figsize=(6.5, 5))
    bars = plt.bar(terms, frequencies, color='skyblue')
    plt.xlabel('Sensors')
    plt.ylabel('Frequency')
    plt.title(title)
    plt.xticks(rotation=45, ha="right")
    
    # Adiciona porcentagens acima das barras
    total = sum(frequencies)
    for bar in bars:
        height = bar.get_height()
        plt.annotate(f'{height/total:.1%}', 
                     xy=(bar.get_x() + bar.get_width() / 2, height), 
                     xytext=(0, 3),  # 3 points vertical offset
                     textcoords="offset points",
                     ha='center', va='bottom')
    
    plt.tight_layout()
    plt.show()

```

Aqui, na lista personalizável, podem ser colocadas tipos de sensores, ou, nomes de satélites por exemplo: 

```python
# Lista de termos em inglês
sensor_terms = ['radar', 'lidar', 'optical']  # Personalizável

# Aplicar a função para criar o gráfico de barras
generate_bar_chart(df_clean, 'Preprocessed_Title', sensor_terms, 'Sensor Terms')
```

Métricas para avaliar a quantidade de tópicos que serão utilizados para Modelagem de Tópicos:

```python
# Bibliotecas para o PLN
#!pip install gensim -q
import gensim
from gensim import corpora
from gensim.models import CoherenceModel, LdaModel
#from gensim.models import LdaModel
from gensim.parsing.preprocessing import preprocess_string, remove_stopwords
```

```python
# Título e resumo em um único campo de texto
df_clean['Text'] = df_clean['Preprocessed_Title'] + ' ' + df_clean['Preprocessed_Abstract']

# Pré-processamento adicional de texto
custom_stopwords = set(stopwords.words('english'))  
df_clean['Text'] = df_clean['Text'].apply(lambda x: remove_stopwords(x))
df_clean['Text'] = df_clean['Text'].apply(lambda x: preprocess_string(x))

# Dicionário a partir do texto pré-processado
dictionary = corpora.Dictionary(df_clean['Text'])

# Corpus 
corpus = [dictionary.doc2bow(text) for text in df_clean['Text']]

# Avaliação de vários modelos com diferentes números de tópicos
start_topics, end_topics, step = 3, 20, 2
num_topics_range = range(start_topics, end_topics + 1, step)
perplexity_scores = []
coherence_scores = []

for num_topics in num_topics_range:
    model = LdaModel(corpus=corpus, num_topics=num_topics, id2word=dictionary, passes=15)
    perplexity = model.log_perplexity(corpus)
    coherence_model_lda = CoherenceModel(model=model, texts=df_clean['Text'], dictionary=dictionary, coherence='c_v')
    coherence_lda = coherence_model_lda.get_coherence()
    
    perplexity_scores.append(perplexity)
    coherence_scores.append(coherence_lda)

    print(f"Num Tópicos: {num_topics}, Perplexidade: {perplexity}, Coerência: {coherence_lda}")
```

```python
# Plotagem dos gráficos
fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('Number of Topics')
ax1.set_ylabel('Perplexity', color=color)
ax1.plot(num_topics_range, perplexity_scores, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('Coherence', color=color)
ax2.plot(num_topics_range, coherence_scores, color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()
plt.title('LDA Model Assessment')
plt.show()
```

### Técnicas relacionadas à Detecção de Mudanças em Sensoriamento Remoto com base no conteúdo dos artigos:

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
```

```python
start_time = time.time()

#np.random.seed(42)

# Concatena os títulos e resumos de cada artigo
df_clean['Title_Abstract'] = df_clean['Preprocessed_Title'] + ' ' + df_clean['Preprocessed_Abstract']

# Vetor de contagem de palavras
vectorizer = CountVectorizer(max_df=0.85, max_features=2000, stop_words='english')
X = vectorizer.fit_transform(df_clean['Title_Abstract'])

# Modelo de tópicos LDA
lda = LatentDirichletAllocation(n_components=10, random_state=42)
lda.fit(X)

end_time = time.time()    # Registra o tempo final
execution_time = end_time - start_time  # Calcula o tempo de execução

print(f"Tempo de execução: {execution_time} segundos")
```

```python
#np.random.seed(42)

# Tópicos e palavras-chave principais para cada tópico
print("Tópicos identificados pelo LDA:")
for topic_idx, topic in enumerate(lda.components_):
    print(f"Tópico {topic_idx + 1}:")
    top_words_idx = topic.argsort()[-10:][::-1]
    top_words = [vectorizer.get_feature_names_out()[i] for i in top_words_idx]
    print(', '.join(top_words))
    print()

# Associação de artigos aos tópicos dominantes
df_clean['Dominant_Topic'] = lda.transform(X).argmax(axis=1)
```

```python
# Extrair países de afiliações
from country_list import countries_for_language
import networkx as nx
countries = dict(countries_for_language('en'))
countries_list = list(countries.values())

# extrai os países de cada linha da afiliação
def extrair_paises(dataFrame):
    retorno = []
    for pais in countries_list:
        if pais in dataFrame:
            retorno.append(pais)
    return retorno

# Grafo usando NetworkX
collaboration_network = nx.Graph()

# Extrai países de afiliações em df_clean, uma por uma
conexoes_paises = []
for i in df_clean.index: 
    paises_por_publicacao = extrair_paises(df_clean['Affiliations'][i])
    # Lista de conexões entre países
    for pais1 in paises_por_publicacao:
        for pais2 in paises_por_publicacao:
            if pais1 != pais2:
                conexoes_paises.append((pais1, pais2))
                collaboration_network.add_edge(pais1, pais2)
```

```python
# Contagem de publicações por país
country_publication_counts = df_clean['Affiliations'].apply(extrair_paises).explode().value_counts()

# Quantidade desejada de países
top_countries_count = 1  # Personalizável

# Países com mais publicações
top_countries = country_publication_counts.head(top_countries_count).index

# Filtra o DataFrame original para incluir apenas as linhas com os países selecionados
df_filtered = df_clean[df_clean['Affiliations'].apply(lambda x: any(country in x for country in top_countries))]
```

```python
start_time = time.time()

# Gráfico de colaboração com base nos países filtrados
collaboration_network_filtered = nx.Graph()
conexoes_paises_filtered = []

# Extrai países de afiliações em df_filtered, uma por uma
for i in df_filtered.index:
    paises_por_publicacao = extrair_paises(df_filtered['Affiliations'][i])
    # Lista de conexões entre países
    for pais1 in paises_por_publicacao:
        for pais2 in paises_por_publicacao:
            if pais1 != pais2:
                conexoes_paises_filtered.append((pais1, pais2))
                collaboration_network_filtered.add_edge(pais1, pais2)

# Layout do gráfico
pos = nx.spring_layout(collaboration_network_filtered, seed=42)

# Tamanho do nó proporcional à sua centralidade
node_size = [collaboration_network_filtered.degree(node) * 50 for node in collaboration_network_filtered.nodes]

# Grafo de colaboração
plt.figure(figsize=(10, 10))
nx.draw(collaboration_network_filtered, pos, node_size=node_size, with_labels=True, font_size=10, node_color='skyblue')
#plt.title(f'Rede de Colaboração entre Países (Top {top_countries_count})')
plt.show()

end_time = time.time()    # Registra o tempo final
execution_time = end_time - start_time  # Calcula o tempo de execução

print(f"Tempo de execução: {execution_time} segundos")
```

```python
# Palavras-chave em inglês para buscar nos documentos (é personalizável)
keywords = ['brazil']

# Função para encontrar documentos com as palavras-chave
def find_documents_with_keywords(df, keywords):
    matching_documents = []

    for _, row in df.iterrows():
        abstract = row['Preprocessed_Abstract']
        title = row['Preprocessed_Title']

        # Verificando se as palavras-chave estão presentes no abstract ou no title
        for keyword in keywords:
            if keyword in abstract or keyword in title:
                matching_documents.append((row['Title'], row['Abstract']))

    return matching_documents

# Encontra documentos com as palavras-chave
matching_documents = find_documents_with_keywords(df_clean, keywords)

# Contagem total de documentos correspondentes
total_matching_documents = len(matching_documents)
```

```python
from gensim import corpora, models
```

```python
# Pré-processamento dos documentos encontrados com base na palavra-chave escolhida
documents = [doc[1] for doc in matching_documents]  # Seleciona apenas os resumos

# Carrega as stopwords em inglês
stop_words = set(stopwords.words('english'))

# Função para pré-processar um texto
def preprocess_text(text):
    # Tokenização
    words = word_tokenize(text)

    # Remoção de stopwords e caracteres especiais
    words = [word.lower() for word in words if word.isalpha() and word.lower() not in stop_words]

    # Lematização
    lemmatizer = WordNetLemmatizer()
    words = [lemmatizer.lemmatize(word) for word in words]

    # Reconstruir o texto pré-processado
    preprocessed_text = ' '.join(words)

    return preprocessed_text

# Pré-processar os documentos
preprocessed_documents = [preprocess_text(doc) for doc in documents]

# Dicionário de palavras
dictionary = corpora.Dictionary(doc.split() for doc in preprocessed_documents)

# Corpus (representação vetorial dos documentos)
corpus = [dictionary.doc2bow(doc.split()) for doc in preprocessed_documents]

# Modelo de tópicos (LDA)
num_topics = 3  # Número de tópicos a serem identificados (ajuste conforme necessário)
lda_model = models.LdaModel(corpus, num_topics=num_topics, id2word=dictionary, passes=45, 
                            random_state=45)
```

```python
# Termos mais frequentes em cada tópico
print("\nTermos mais frequentes em cada tópico:")
topics = lda_model.print_topics(num_words=8)  # Altera o número de termos para cada tópico conforme necessário
for topic in topics:
    print(topic)
```
